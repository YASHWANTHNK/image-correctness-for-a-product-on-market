```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
def load_and_preprocess_image(img_path):
    """
    Load an image and preprocess it for the model.
    """
    img = tf.io.read_file(img_path)  # Read the image file
    img = tf.io.decode_jpeg(img, channels=3)  # Decode the image
    img = tf.image.resize(img, size=(384, 384), antialias=True)  # Resize the image to match model input size
    img = img / 255.0  # Normalize the image (scaling from 0 to 1)
    img = tf.expand_dims(img, axis=0)  # Add a batch dimension (convert 3D to 4D)
    return img
def predict_dehazed_image(model_path, img_path):
    """
    Load the saved model and make predictions on the input image.
    """
    # Load the saved model
    model = tf.keras.models.load_model(model_path, compile=False)

    # Load and preprocess the input image
    input_img = load_and_preprocess_image(img_path)

    # Perform prediction
    dehazed_img = model(input_img, training=False)

    # Display the original hazy image and the dehazed output
    plt.figure(figsize=(10, 5))
    display_list = [input_img[0], dehazed_img[0]]
    titles = ['Hazy Image', 'Dehazed Image']

    for i in range(2):
        plt.subplot(1, 2, i + 1)
        plt.title(titles[i])
        plt.imshow(display_list[i])
        plt.axis('off')

    plt.show()
#trained_model
# Example usage
model_path = r'model/trained_model'  # Use the directory, not the specific file
img_path = r'images/002677.jpg'

predict_dehazed_image(model_path, img_path)
```
